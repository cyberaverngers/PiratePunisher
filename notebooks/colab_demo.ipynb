```python
# Install dependencies (will install pytest as a dev dependency for tests)
!git clone --depth 1 https://github.com/cyberaverngers/PiratePunisher.git repo
import requests, os

# Prompt for a GitHub Personal Access Token (PAT) with repo scope
token = getpass('GitHub PAT (with repo scope): ')
headers = {'Authorization': f'token {token}'}
url = 'https://api.github.com/repos/cyberaverngers/PiratePunisher/zipball/main'

r = requests.get(url, headers=headers)
if r.status_code == 200:
    open('repo.zip','wb').write(r.content)
    print('Downloaded repo zip, extracting...')
    !unzip -q repo.zip
    # find extracted dir and rename to 'repo'
    extracted = [d for d in os.listdir('.') if d.endswith('PiratePunisher-main') or d.startswith('cyberaverngers-PiratePunisher-')]
    if extracted:
        os.rename(extracted[0], 'repo')
        print('Repo extracted to ./repo')
    else:
        print('Could not find extracted directory; listing current dir:')
        !ls -la
else:
    print('Failed to download repo:', r.status_code, r.text)

# Clear token from memory
del token
```
